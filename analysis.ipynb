```python
import pandas as pd
import json
from collections import defaultdict
```

```python
df = pd.read_csv("result_summary.csv")
print(len(df))
print(df.head(3))
```

```python
with open("personas_40_selected.json", "r") as f: 
    persona = json.load(f)
```

```python
threshhold = 2
df_filtered = df.query(f"opinion_improvement > {threshhold} and morale_improvement > {threshhold} and website_improvement > {threshhold}")
print(len(df_filtered))

lst_good_cases = df_filtered["persona_id"]

print(lst_good_cases)
```

```python

def print_selected_personas(lst_persona):

    att_dict = defaultdict(set)

    for idx, p in enumerate(persona):
        if idx+1 in lst_persona:
            opa_filter = p["opa_filters"]
            for k,v in opa_filter.items():
                att_dict[k].add(v)
            print(f"Persona: {idx+1}")
            print(p["description"])
            print("-"*30)

    for k, v in att_dict.items():
        print(f"{k} : {v}")
    
    
```

```python
print("List of personas with more than 5% gain in all dataset:")
print_selected_personas(lst_good_cases)
```

```python
threshhold = 5
base_threshold = 0.3
df_filtered2 = df.query(f"opinion_improvement > {threshhold} and morale_improvement > {threshhold} and website_improvement > {threshhold} and morale_baseline>{base_threshold} and website_baseline>{base_threshold} and opinion_baseline>{base_threshold}")
print(len(df_filtered2))
lst_good_cases_v2 = df_filtered2["persona_id"]
print("List of personas with more than 5% gain and strong baseline (>0.3) in all dataset:")
print_selected_personas(lst_good_cases_v2)
```

```python
threshhold = -0.5
df_bad = df.query(f"opinion_improvement < {threshhold} and morale_improvement < {threshhold} and website_improvement < {threshhold}")
print(len(df_bad))
lst_bad_cases = df_bad["persona_id"]
print(lst_bad_cases)
print_selected_personas(lst_bad_cases)
```

```python
threshhold = -0.5
base_threshold = 0.3
df_filtered2 = df.query(f"opinion_improvement < {threshhold} and morale_improvement < {threshhold} and website_improvement < {threshhold} and morale_baseline>{base_threshold} and website_baseline>{base_threshold} and opinion_baseline>{base_threshold}")
print(len(df_filtered2))
lst_good_cases_v2 = df_filtered2["persona_id"]
print("List of personas with less than -0.5% gain and strong baseline (>0.3) in all dataset:")
print_selected_personas(lst_good_cases_v2)
```
